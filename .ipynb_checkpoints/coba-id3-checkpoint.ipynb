#  Import library and data

```python
# import library and data
import numpy as np 
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
%matplotlib inline
import math
import seaborn as sns

df = pd.read_csv("kag_risk_factors_cervical_cancer.csv")
```

# Cleaning

## Null Handling

```python
df.info()
```

```python
df[df == '?'].count()
```

```python
df.drop(['STDs: Time since first diagnosis','STDs: Time since last diagnosis'],inplace=True,axis=1)
```

```python
df[df == '?'].count()
```

### Fill NaN Numerical Data with Median

```python
numerical = ['Age', 'Number of sexual partners', 'First sexual intercourse','Num of pregnancies', 'Smokes (years)', 
             'Smokes (packs/year)','Hormonal Contraceptives (years)','IUD (years)','STDs (number)']
```

```python
df = df.replace('?', np.NaN)
```

```python
for feature in numerical:
    print(feature,'',df[feature].astype(str).astype(float).mean())
    feature_mean = round(df[feature].astype(str).astype(float).mean(),1)
    df[feature] = df[feature].fillna(feature_mean)
```

```python
df.isna().sum()
```

### Fill NaN Categorical Data with Mode

```python
categorical = ['Smokes', 'Hormonal Contraceptives', 'IUD','STDs', 'STDs:condylomatosis', 
               'STDs:cervical condylomatosis','STDs:vaginal condylomatosis','STDs:vulvo-perineal condylomatosis','STDs:syphilis',
              'STDs:pelvic inflammatory disease', 'STDs:genital herpes', 'STDs:molluscum contagiosum', 'STDs:AIDS', 'STDs:HIV',
              'STDs:Hepatitis B', 'STDs:HPV']
```

```python
df[categorical]=df[categorical].fillna(df.mode().iloc[0])
```

```python
df.head()
```

```python
df.isna().sum()
```

### Drop Duplicate Data

```python
df.drop_duplicates(inplace=True)
```

## Correcting Data Type

```python
for feature in numerical:
    df[numerical] = df[numerical].astype(float)
```

```python
for feature in categorical:
    df[categorical] = df[categorical].astype(object)
```

## Age

```python
def age_classification(n):
    if n<=12 :
        return 0 #Children
    elif n<=17:
         return 1 #Adolescents
    elif n<=44:
         return 2 #Pre-menopausal Adults 
    elif n<=55:
         return 3 #Menopausal Adults
    elif n<=64:
         return 4 #Post-menopausal Adults
    else:
        return 5 #Older Adults
```

```python
df['age_classification'] = df['Age'].apply(age_classification)
df.loc[df['age_classification'] == 3]
```

## Number of sexual partners

```python
def Number_of_sexual_partners_classification(n):
    if n<=2 :
        return 0 #Few
    elif n<=3:
         return 1 #Several
    else:
        return 2 #Many
```

```python
df['Number_of_sexual_partners_classification'] = df['Number of sexual partners'].apply(Number_of_sexual_partners_classification)
df.loc[df['Number_of_sexual_partners_classification'] == 1]
```

## First sexual intercourse

```python
def First_sexual_intercourse_classification(n):
    if n<=15 :
        return 0 #Young
    elif n<=18:
         return 1 #Not too young
    else:
        return 2 #Old
```

```python
df['First sexual intercourse classification'] = df['First sexual intercourse'].apply(First_sexual_intercourse_classification)
df.loc[df['First sexual intercourse classification'] == 2]
```

## Num of pregnancies

```python
def Num_of_pregnancies_classification(n):
    if n<=1 :
         return 0 #Few
    elif n<=3:
         return 1 #Several
    else:
         return 2 #Many
```

```python
df['Num of pregnancies classification'] = df['Num of pregnancies'].apply(Num_of_pregnancies_classification)
df.loc[df['Num of pregnancies classification'] == 1]
```

## Smokes (years)

```python
def Smokes_years_classification(n):
    if n<=1 :
        return 0 #New
    else:
        return 1 #Old
```

```python
df['Smokes (years) classification'] = df['Smokes (years)'].apply(Smokes_years_classification)
df.loc[df['Smokes (years) classification'] == 1]
```

## Smokes (packs/year)

```python
def Smokes_packsperyear_classification(n):
    if n<=1 :
        return 0 #Few
    else:
        return 1 #Many
```

```python
df['Smokes (packs/year) classification'] = df['Smokes (packs/year)'].apply(Smokes_packsperyear_classification)
df.loc[df['Smokes (packs/year) classification'] == 1]
```

## Hormonal Contraceptives (years)

```python
def Hormonal_Contraceptives_years_classification(n):
    if n<=1 :
        return 0 #New
    elif n<=3:
        return 1 #Not_long_ago
    else:
        return 2 #Old
```

```python
df['Hormonal Contraceptives (years) classification'] = df['Hormonal Contraceptives (years)'].apply(Hormonal_Contraceptives_years_classification)
df.loc[df['Hormonal Contraceptives (years) classification'] == 0]
```

## IUD (years)

```python
def IUD_years_classification(n):
    if n<=2 :
        return 0 #New
    else:
        return 1 #New
```

```python
df['IUD (years) classification'] = df['IUD (years)'].apply(IUD_years_classification)
df.loc[df['IUD (years) classification'] == 0]
```

## STDs (number)

```python
def STDs_number_classification(n):
    if n<=1 :
        return 0 #Few
    else:
        return 1 #Many
```

```python
df['STDs (number) classification'] = df['STDs (number)'].apply(STDs_number_classification)
df.loc[df['STDs (number) classification'] == 0]
```

# Labelling

Pada data, terdapat 4 tes untuk menentukan apakah pasien menderita kanker serviks atau tidak, yaitu Hinselmann, Schiller, Citology, dan Biopsy.

Untuk percobaan ini, akan dibuat label seperti berikut:
* Jika Hinselmann + Schiller + Citology + Biopsy = 0 maka Prediction = 0
* Jika Hinselmann + Schiller + Citology + Biopsy > 0 maka Prediction = 1

```python
df.head()
```

```python
df.tail()
```

```python
def label(h, s, c, b):
    if h+s+c+b == 0 :
        return 0
    else:
        return 1
```

```python
df['Prediction'] = df[['Hinselmann', 'Schiller', 'Citology', 'Biopsy']].apply(lambda row: label(*row), axis=1)
df.loc[df['Prediction'] == 0]
```

# Drop label lama

```python
df.drop(['Hinselmann','Schiller', 'Citology', 'Biopsy'],inplace=True,axis=1)
df.head()
```

# Check Correlation

```python
corr = df2.corr()
```

```python
plt.figure(figsize=(50,30))
sns.heatmap(corr, cmap=plt.cm.CMRmap_r,annot=True)
plt.show()  
```

```python
# corr_features = correlation(df2, 0.85)
# corr_features
```

```python
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
```

```python
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]
to_drop
```

```python
df_corr85 = df2.copy()
```

```python
df_corr85.drop(to_drop, axis=1, inplace=True)
df_corr85.head()
```

```python
y = df_corr85['Prediction']
X = df_corr85.drop(['Prediction'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Split the dataset.

X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.2, random_state=30)

# Fit data. 

clf = DecisionTreeClassifier(max_depth=8, min_samples_split=150)
clf = clf.fit(X_train, y_train)

# Check classifier accuracy.

accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)

# Plot the tree.

feature_names = ["age_classification", "Number_of_sexual_partners_classification", "First sexual intercourse classification", "Num of pregnancies classification", "Smokes (years) classification", "Smokes (packs/year) classification", "Hormonal Contraceptives (years) classification", "IUD (years) classification", "STDs (number) classification", "STDs_condylomatosis", "STDs_vaginal_condylomatosis", "STDs_vulvo_perineal_condylomatosis", "STDs_syphilis", "STDs_pelvic_inflammatory_disease", "STDs_genital_herpes", "STDs_molluscum_contagiosum", "STDs_HIV", "STDs_Hepatitis_B", "STDs_HPV", "STDs: Number of diagnosis", "Dx:Cancer", "Dx:CIN", "Dx:HPV"]
class_names=["not predicted" "predicted"]
plt.figure(figsize=(40,30))
tree.plot_tree(clf,feature_names=feature_names, class_names=["not predicted", "predicted"],filled=True,rounded=True)
plt.show()
```
